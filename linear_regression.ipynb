# Linear Regression model

## $y = \beta_1 x + \beta_0$
### where, $\beta_1$ and $\beta_0$ are model parameters.
### x is independent (and the feature) and y is dependent variable

## Error calculation $\epsilon$ is through Mean Squared Error 
## $ \epsilon = \frac{1}{n}\Sigma^n_{i=1}(\beta_1 x + \beta_0 - y_i)^2 $

```python
#import dependencies
import numpy as np 
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
```

```python
#data preprocessing
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
# end of preprocessing

#start of model code and training
model = LinearRegression().fit(x,y)
# end of training code

#output
z_score = model.score(x,y)
print(f"z-score: {z_score}")
print(f"parameter beta-naught {model.intercept_}")
print(f"parameter beta-one {model.coef_}\n")

# y_pred = model.predict(x)
```

### Linear Regression uses an algorithm called "Gradient Descent" to minimise parameter values to minimise errors

## Trained! Now predicting values

```python
y_pred = model.coef_*x + model.intercept_
```

```python

print("ORIGINAL VALUES -------------------->\n")

for i in range(len(x)):
    print(f"({x[i][0]},{round(y[i],2)})")


for i in range(len(x)):
    plt.scatter(x[i][0],y[i])

print("\nPREDICTED VALUES -------------------->\n")
y_pred = (model.coef_ * x) + model.intercept_ #ax + b
for i in range(len(x)):
    print(f"({x[i][0]},{round(y_pred[i][0],1)})")    
    
xpl = np.reshape(x,(1,len(x)))
ypl = np.reshape(y_pred,(1,len(y)))
plt.plot(xpl[0],ypl[0])

print("\nORIGINAL GRAPH (Scatter) & PREDICTED (Best fit)")
```
